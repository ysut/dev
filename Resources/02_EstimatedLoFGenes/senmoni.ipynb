```python

# 1. 症例リストの作成
# detailed_diseases = [ ... ]  # 詳細な診断名のリスト
cases = [(random.choice(detailed_diseases), np.random.randint(18, 65)) for _ in range(50)]

# 2. 診療日の追加
def random_date(start_date, end_date):
    delta = end_date - start_date
    random_days = random.randrange(delta.days)
    return start_date + timedelta(days=random_days)

start_date = datetime.strptime("2019-05-01", "%Y-%m-%d")
end_date = datetime.strptime("2024-03-31", "%Y-%m-%d")
cases_with_dates = [(case[0], case[1], random_date(start_date, end_date)) for case in cases]
sorted_cases_with_dates = sorted(cases_with_dates, key=lambda x: x[2])

# 3. 検査・処置・治療の付与
# interventions = [ ... ]  # 検査・処置・治療の例
cases_with_interventions = [(case[2], case[1], case[0], random.choice(interventions)) for case in sorted_cases_with_dates]

# 4. リストの整形と出力
for case in cases_with_interventions:
    print(case[0].strftime("%Y-%m-%d"), case[1], "歳,", case[2] + ",", case[3])

```
